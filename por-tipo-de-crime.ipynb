```python
# Importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

# Lê a tabela CSV
df = pd.read_csv('Crimes_brasil_uf.csv')

# Agrupa os dados por Tipo_crime e conta o número de ocorrências
crime_count = df.groupby('Tipo_crime').size().reset_index(name='Count')

# Cria o gráfico de barras
crime_count.plot.bar(x='Tipo_crime', y='Count', rot=0)

# Adiciona título e legendas ao gráfico
plt.title("Número de ocorrências de crimes no Brasil - 2015/2022")
plt.xlabel(" ")
plt.ylabel("Número de ocorrências")
plt.xticks(rotation=45, ha='right', fontsize=10)

# Exibe o gráfico
plt.show()
```
